# Limits

The Calculus of Infinitesimals

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000) # start, finish, n points
```

```python
y = x**2 + 2*x + 2
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim(-1.01, -0.99)
ax.set_ylim(0.99, 1.01)
_ = ax.plot(x, y)
```
