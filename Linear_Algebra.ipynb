Norms are functions that quantify vector magnitude

L<sup>2</sup> Norm OR Simple Euclidean distance from origin 

$
\| \mathbf{x} \|_2 = \sqrt{\sum_i x_i^2}
$

```python
x
```

```python
np.linalg.norm(x)
```

Unit Vector

If ||x|| = 1, then X is a unit vector

```python
x = np.array([1,0,0])
np.linalg.norm(x)
```

L<sup>1</sup> Norm

$
\| \mathbf{x} \|_1 = \sum_i |x_i|\
$

Varies linearly at all locations whether near or far from origin.  
Used when difference between zero and non-zero is key.

```python
np.linalg.norm(x, ord= 1)
```

Squared L<sup>2</sup> Norm OR Euclidean norm squared

$
\|x\|_2^2 = \sum_i x_i^2
$

It is computationally cheaper as it is equat to the dot product of the vector.  
Derivatives of element x requires that element alone, whereas L2 norm reequires entire vector.
It grows very slowly near the origin. So cannot be used where diff between zero and near-zero is important.

```python
x
```

```python
np.dot(x, x)
```

Max OR Infinite Norm

$
\|\mathbf{x}\|_\infty = \max_i |x_i|
$

```python
np.linalg.norm(x, ord=np.inf)
```

Generalized LP Norm

p must be a real number
p must be >= 1

$
\|\mathbf{x}\|_p = \left( \sum_{i} |x_i|^p \right)^{\frac{1}{p}}
$

Unit vectors are used as basis vectors. Basis vectors can be scaled to represent any vector in given vector space.

Orthogonal vectors

x and y are orthogonal vectors if:  
$
\mathbf{x}^{\mathrm{T}}\mathbf{y} = 0
$


Assuming non-zero norms:  
    They are at 90 degrees to each other  
    n-dimentional space has max n mutually orthogonal vectors.

Orthonormal vectors are orthogonal vectors all having unit norm

```python
i = np.array([1, 0])
j = np.array([0, 1])

np.dot(i, j)
```

Matrix Inversion

Matrix Inverse of $ X $ is $ X^{-1} $

$ X^{-1} X  = I_{n} $

$
y = Xw\\  
Xw = y\\
X^{-1} Xw = X^{-1} y\\
I_n w = X^{-1} y\\
w = X^{-1} y\\
$

```python
X = np.array([[4,2],[-5,-3]])
X
```

```python
Xinv = np.linalg.inv(X)
Xinv
```

```python
y = np.array([4,-7])
y
```

```python
w = np.dot(Xinv,y)
w
```

Applying a matrix to a vector can linearly transform the vector. Identity matrix is an exception.

```python
v = np.array([3,1])
v
```

```python
import matplotlib.pyplot as plt
```

```python
def plotVectors(vecs, cols, alpha = 1):
    plt.figure()
    plt.axvline(x=0, color='#A9A9A9', zorder = 0)
    plt.axhline(y=0, color='#A9A9A9', zorder = 0)

    for i in range(len(vecs)):
        x = np.concatenate([[0,0], vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i],
                   alpha=alpha)
```

```python
plotVectors([v], cols=['lightblue'])
ax = plt.gca()
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
```

```python
I = np.array([[1,0],[0,1]])
I
```

```python
Iv = np.dot(I,v)
Iv
```

```python
Iv == v
```

```python
plotVectors([Iv], cols=['blue'])
ax = plt.gca()
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
```

```python
A = np.array([[-1,4],[2,-2]])
A
```

```python
Av = np.dot(A,v)
```

```python
plotVectors([v, Av], cols = ['lightblue', 'blue'])
ax = plt.gca()
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
```

```python
v2 = np.array([2,1])
plotVectors([v2, np.dot(A, v2)], cols= ['lightgreen', 'green'])
ax = plt.gca()
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
```
